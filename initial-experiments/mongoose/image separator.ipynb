```python
def extract_sprites_horizontal(sprite_sheet_paths, num_frames, output_dir , offset, FLIP_LEFT_RIGHT = False):
    """
    Extracts individual sprites from a horizontally aligned sprite sheet and saves them as separate images.

    Parameters:
    sprite_sheet_path (str): Path to the sprite sheet image.
    num_frames (int): Total number of frames (sprites) in the sprite sheet.
    output_dir (str): Directory where extracted sprites will be saved.
    """
    global numbering
    # Open the sprite sheet
    for i in range(len(sprite_sheet_paths)):
        sprite_sheet = Image.open(sprite_sheet_paths[i])

        # Get sprite sheet dimensions
        sheet_width, sheet_height = sprite_sheet.size

        # Calculate the width and height of each frame
        frame_width = sheet_width // num_frames
        frame_height = sheet_height  # Height of each frame is the same as the height of the sprite sheet

        # Ensure output directory exists
        os.makedirs(output_dir, exist_ok=True)
        r = random.randint(0, 2)
        off = [random.randint(offset[0][0], offset[0][1]) ,random.randint(offset[1][0], offset[1][1])]
        # Loop through each frame and extract it
        for frame_num in range(num_frames):
            # Calculate the coordinates of the sprite in the sprite sheet
            left = frame_num * frame_width
            upper = 0
            right = left + frame_width
            lower = frame_height

            # Crop the sprite from the sprite sheet
            frame = sprite_sheet.crop((left, upper, right, lower))
            # 새로운 빈 이미지 생성 (원래 이미지와 동일한 크기, 투명 배경)
            new_frame = Image.new("RGBA", frame.size, (255, 255, 255, 0))

            # frame 이미지를 상하좌우로 랜점하게 이동하여 new_frame에 붙여넣기
            new_frame.paste(frame, off)
            # new_frame 의 이미지를 좌우 반전 시켜서 저장
            if FLIP_LEFT_RIGHT == True :
              if r== 1:
                new_frame = new_frame.transpose(Image.FLIP_LEFT_RIGHT)

            # Save the frame as a separate image file
            new_frame.save(f"{output_dir}/frame_{numbering}_{frame_num}.png")

        print(f"{num_frames} frames extracted and saved in {output_dir}")
        numbering += 1
```

```python
def make_npy(output_name, sheet_paths) :
  dir = '/content/drive/MyDrive/NIPA/a-yo-image/initial-experiments/mongoose/separator_output/' + output_name + '/'
  data_stack = []
  for j in range(len(sheet_paths)) :
    img_stack = []
    for i in range(10) :
    # 이미지 파일 불러오기
      img = Image.open(dir + 'frame_'+ str(j) +'_' + str(i) + '.png')

      # 이미지를 (h, w, c) 형식의 NumPy 배열로 변환
      img_stack.append(img)
    img_stack = np.stack(img_stack)
    data_stack.append(img_stack)

  data_stack = np.stack(data_stack)
  print("Image array shape:", data_stack.shape)

  np.save('./data/' + output_name + '.npy', data_stack)
```
